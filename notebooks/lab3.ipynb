MIPS-анализ: «эко-эффективность» - получение большого количества продуктов или услуг с наименьшими издержками с минимальными расходованием всех видов ресурсов (или с их максимально эффективным использованием : снижение потребления природных ресурсов (рост эко-эффективнсти )- внедрение новых технологий и в меньшей степени устранение недостатков в этой части существующих технологий.
MIPS-анализ основан на концепции «дематериализаци».Этот подход предусматривает переход от рассмотрения экологически опасных «выходов» к анализу «входов» отдельных процессов или продуктовых цепей в целом. ( Расчёт показал ,что реализация данного подхода позволит  10 раз сохранить потребление природных ресурсов ,необходимых для обеспечения населения стран Западной Европы. Для стального мира увеличение эффективности использования природных ресурсов позволяет сохранить в 2 раза на входе ,повысить уровень жизни в 2 раза. Это «фактор 10» и «фактор 4»). Уровень дематериализации может быть измерен в единицах MIPS-Material Input Per Unit Service or Utility.
MIPS в переводе означает «материальный вход на единицу услуги или полезного продукта». «Полезный» означает, что продукт имеет разную ценность. MIPS показывает, какое количество природных ресурсов (материалов) используется для получения данного полезного продукта или услуги MIPS=MI/S, где MI-количество материалов на «входе»процесса или продуктовой цепи; S – количество оказываемых услуг или выпускаемого полезного продукта.
S/MI – природно-ресурсная эффективность или эко-эффективность. MIPS-анализ позволяет определить расход природных ресурсов на границе продуктовой цепи, в месте их извлечения из природной среды, а также на всём протяжении жизненного цикла продукта или услуг.
Расчёты производятся исходя из количества тонн материалов, извлекаемых из природы. Отдельно рассматриваются биотические, или возобновляемые, природные ресурсы и абиотические, или не возобновляемые, природные ресурсы, а также вода, атмосферный воздух и эрозия почвенно-земельных ресурсов. Потребляемые ресурсы во время производства, использования и рециклирования отходов продукта пересчитываются с помощью специальных переводных коэффициентов или MI- чисел в количество используемых природных ресурсов. (Специалистами Германии Wuppertal Institute определены значения чисел MI не только для ряда материалов, но и для определённых видов энергии и транспорта, которые соответственно имеют размерность т/кВт-ч и т/т-км – http:\\www.mips-online.com)
Согласно концепции MIPS любой продукт несёт невидимый «экологический рюкзак». Он равен: (экологический рюкзак) = (MI)-(вес полезного продукта).
